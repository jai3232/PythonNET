## Question 1 - Function
Create a recursive function that returns the n_th number of element in a Fibonacci sequence.
A Fibonacci sequence is a series of numbers whereby the number is the sum of the two preceding ones in the sequence.

[0,1,1,2,3,5,8,13,21...]

*Note: A recursive function is a function that returns the function itself.

````python
# Answer
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(5))
print(fibonacci(10))
````

```python
list_1 = [11, 2, 6, 12, 15, 18, 21]
list_2 = [32, 7, 12, 16, 9, 27, 88]
```

Given two lists, list_1 and list_2, write a program to create a third list, list_3, by picking all the odd-numbers from list_1 and even-numbers from list_2. Your code should still work even if list_1 and list_2 has changed.

Expected result:
```python
list_3 = [11,15,21,32,12,16,88]
```

````python
list_1 = [11, 2, 6, 12, 15, 18, 21]
list_2 = [32, 7, 12, 16, 9, 27, 88]


odd_arr = []
even_arr = []
for num in list_1:
    if num % 2 != 0:
        odd_arr.append(num)
for num in list_2:
    if num % 2 == 0:
        even_arr.append(num)
list_3 = odd_arr + even_arr

print(list_3)
````

### Part D
Override the `introduce_yourself()` method, such that now it prints a string with the following format: "Hi, my name is (person's name), I am a/an (person's job), currently working in (person's workplace)."

### Part E
Create two objects called "human_1" and "human_2", using the "Person" and "Employee" class respectively. You may create the objects with your own desired values. Then, let the two humans introduce themselves by calling the `introduce_yourself()` method respectively.


````python
# Answer part A
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

# Answer part B
    def introduce_yourself(self):
        return f"Hi, my name is {self.name}, I am {self.age} years old."

boy = Person("John", 11, "M")
print(boy.introduce_yourself())

# Answer part C
class Employee(Person):
    def __init__(self, name2, age, gender, job, salary, workplace):
        super().__init__(name2, age, gender)
        self.job = job
        self.salary = salary
        self.workplace = workplace

# Answer part D
    def introduce_yourself(self):
        return f"Hi, my name is {self.name}, I am a/an {self.job}, currently working in {self.workplace}."
doctor = Employee("Mike", 27, "M", "Docktor", 10000, "London")
print(doctor.introduce_yourself())

# Answer part E
human_1 = Employee("Sarah", 33, "F", "Sale Girl", 5000, "Singapore")
print(human_1.introduce_yourself())
human_2 = Employee("Vladimir", 33, "M", "engineer", 7000, "Moscow")
print(human_2.introduce_yourself())
````
