## Part1: Matplotlib
### Section 1: Creating Basic Plots
Description: This section covers the creation of basic plots using Matplotlib, including line plots, scatter plots, and bar charts.

Creating Basic Plots:
- plot(): Create line plots.
- scatter(): Create scatter plots.
- bar(): Create bar charts.

```python
import matplotlib.pyplot as plt
import numpy as np

# Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
# Scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
plt.figure()
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### Section 2: Customizing Plots
Description: This section demonstrates how to customize plots by changing colors, markers, and line styles.

Customizing Plots:
- Changing colors.
- Changing markers.
- Changing line styles.

```python
import matplotlib.pyplot as plt
import numpy as np

# Line plot with customization
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.plot(x*x /3, y, color='blue', linestyle='--', marker='o')
plt.title("Customized Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
# Scatter plot with customization
x = np.random.rand(100)
y = np.random.rand(100)
plt.figure()
plt.scatter(x, y, color='green', marker='x')
plt.title("Customized Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Exe: change color to purple, linestyle
import matplotlib.pyplot as plt
import numpy as np

# Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y, color='purple', linestyle="dashed", marker="*")
plt.title("Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

### Section 3: Adding Labels and Titles
Description: This section covers how to add labels, titles, and legends to Matplotlib plots.

Adding Labels and Titles:
- title(): Add a title to the plot.
- xlabel(), ylabel(): Add labels to the x and y axes.
- legend(): Add a legend to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Line plot with labels, title, and legend
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure()
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title("Trigonometric Functions")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
# Create a numpy, x of number from 0 to 10 with 200 elements
# Create y1 which is sin(x)
# Create y2 which is sin(x + 50)
# Plot both graphs with labels: sin(x) and sin(x + 50)
# Title
# xlabel: x, ylabel: value

x = np.linspace(0, 10, 200)
y1 = np.sin(x)
y2 = np.sin(x + 50)
plt.subplot(2, 1, 1)
# plt.plot(x,y)
plt.title("Trigonometric Functions: no Shift")
plt.plot(x, y1, label='sin(x)')
plt.subplot(2, 1, 2)
plt.plot(x, y2, label='sin(x + 50)')
plt.title("Trigonometric Functions: Phase Shift + 50")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.show()
```

### Section 4: Working with Multiple Plots
Description: This section demonstrates how to create multiple plots within the same figure using subplots.

Working with Multiple Plots:
- subplot(): Create multiple plots in the same figure.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()

# First subplot
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title("Sin(x)")

# Second subplot
plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title("Cos(x)")

plt.tight_layout()
plt.show()
```

### Section 5: Creating Advanced Plots
Description: This section covers advanced plots such as histograms, box plots, and heatmaps.

Creating Advanced Plots:
- hist(): Create histograms.
- boxplot(): Create box plots.
- imshow(): Create heatmaps.

```python
import matplotlib.pyplot as plt
import numpy as np

# Histogram
data = np.random.randn(1000)
plt.figure()
plt.hist(data, bins=30)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot
data = [np.random.randn(100) for _ in range(4)]
plt.figure()
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
# Heatmap
data = np.random.rand(10, 10)
plt.figure()
plt.imshow(data, cmap='viridis')
plt.title("Heatmap")
plt.colorbar()
plt.show()
```

### Section 1: Creating Basic Plots
Description: This section covers the creation of basic plots using Seaborn, including line plots, scatter plots, and bar charts.

Creating Basic Plots:
- lineplot(): Create line plots.
- scatterplot(): Create scatter plots.
- barplot(): Create bar charts.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
sns.lineplot(x=x, y=y)
plt.title("Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
# Bar chart
categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.figure()
sns.barplot(x=categories, y=values)
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

### Section 2: Customizing Plots
Description: This section demonstrates how to customize plots by changing colors, markers, and styles using Seaborn.

Customizing Plots:
- Changing colors.

### Section 3: Adding Labels and Titles
Description: This section covers how to add labels, titles, and legends to Seaborn plots.

Adding Labels and Titles:
- set_title(): Add a title to the plot.
- set_xlabel(), set_ylabel(): Add labels to the x and y axes.
- legend(): Add a legend to the plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Line plot with labels, title, and legend
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure()
sns.lineplot(x=x, y=y1, label='sin(x)')
sns.lineplot(x=x, y=y2, label='cos(x)')
plt.title("Trigonometric Functions")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.show()
```

### Section 4: Working with Multiple Plots
Description: This section demonstrates how to create multiple plots within the same figure using Seaborn's FacetGrid and Matplotlib's subplots.

Working with Multiple Plots:
- FacetGrid(): Create multiple plots in the same figure.
- subplot(): Create multiple plots in the same figure.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Multiple plots using FacetGrid
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

```python
# Multiple plots using subplot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()

# First subplot
plt.subplot(2, 1, 1)
sns.lineplot(x=x, y=y1)
plt.title("Sin(x)")

# Second subplot
plt.subplot(2, 1, 2)
sns.lineplot(x=x, y=y2)
plt.title("Cos(x)")

plt.tight_layout()
plt.show()
```

### Section 5: Creating Advanced Plots
Description: This section covers advanced plots such as histograms, box plots, and heatmaps using Seaborn.

Creating Advanced Plots:
- histplot(): Create histograms.
- boxplot(): Create box plots.
- heatmap(): Create heatmaps.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Histogram
data = np.random.randn(1000)
plt.figure()
sns.histplot(data, bins=30, kde=True)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot
import seaborn as sns
data = [np.random.randn(100) for _ in range(4)]
plt.figure()
sns.boxplot(data=data)
plt.title("Box Plot")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
print(np.random.randn(10) for _ in range(4))
```

```python
# Heatmap
data = np.random.rand(10, 10)
plt.figure()
sns.heatmap(data, cmap='viridis')
plt.title("Heatmap")
plt.show()
```

### Matplotlib VS Seaborn

- Matplotlib: Great for **detailed and customized plots**; requires more code and provides extensive control over plot appearance.
- Seaborn: Simplifies the creation of complex statistical plots; offers attractive default aesthetics and integrates well with Pandas for **quick data visualization**.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Value': [1, 2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)
g = sns.FacetGrid(df, col='Category', col_wrap=2)

g.map(plt.hist, 'Value')

g.set_axis_labels('Value', 'Count')

plt.show()
```
